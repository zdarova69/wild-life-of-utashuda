```python
from ultralytics import YOLO
from glob import glob
import torch
import cv2
from sklearn.metrics import mean_squared_error as mse
from PIL import Image
import random
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
import pandas as pd



model = YOLO('/content/ultralytics/runs/detect/train2/weights/best.pt')
model.conf = 25
```

```python
preds = []
y = []

for pth in glob('/content/ultralytics/valid/images/*'):
    img = Image.open(pth)
    results = model(img, stream=True)
    for result in results:
      boxes = result.boxes  # Boxes object for bbox outputs
      preds.append(boxes.xyxy.shape[0])
    f = open(pth.replace('images', 'labels')[:-3] + 'txt', 'r')
    y.append(len(list(f)))
    f.close()

print(f'RMSE = {mse(y, preds, squared=False)**0.5}')
```

```python
paths = []
preds = []
for pth in glob('/content/ultralytics/images/*'):
    img = Image.open(pth)
    results = model(img)
    for result in results:
      boxes = result.boxes  # Boxes object for bbox outputs
      preds.append(boxes.xyxy.shape[0])
    paths.append(pth.split('/')[-1])

df = pd.DataFrame(list(zip(paths, preds)), columns=['filename', 'num'])
df.to_csv('count.csv', index=False)
```
